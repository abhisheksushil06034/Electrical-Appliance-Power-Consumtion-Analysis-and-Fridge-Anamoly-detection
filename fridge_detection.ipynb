```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, LSTM, Dense, Dropout, MaxPooling1D, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Load the data - Updated file path
try:
    df = pd.read_csv('last24h_fridge.csv')  # Removed /content/ path
    print("Data loaded successfully!")
except FileNotFoundError:
    print("Error: File 'last24h_fridge.csv' not found in current directory")
    print("Please ensure the CSV file is in the same directory as this script")
    exit()

# Data preprocessing and validation
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")

# Handle time column parsing with error checking
try:
    df['Time'] = pd.to_datetime(df['Time'])
    df.set_index('Time', inplace=True)
    print("Time column parsed successfully")
except Exception as e:
    print(f"Error parsing time column: {e}")
    # If Time column doesn't exist or can't be parsed, create a datetime index
    df.index = pd.date_range(start='2024-01-01', periods=len(df), freq='T')
    print("Created synthetic time index")

# Check for required Power column
if 'Power' not in df.columns:
    print("Error: 'Power' column not found in dataset")
    print("Available columns:", df.columns.tolist())
    exit()

# Handle missing values
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')
df.dropna(subset=['Power'], inplace=True)
print(f"Data points after cleaning: {len(df)}")

# Plot the data to understand patterns
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['Power'])
plt.title('Fridge Power Consumption - Complete Dataset')
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Data preprocessing
def create_sequences(data, seq_length):
    """Create sequences for time series data"""
    X = []
    for i in range(len(data) - seq_length + 1):  # Fixed off-by-one error
        X.append(data[i:(i + seq_length)])
    return np.array(X)

# Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Power']])

# Sequence parameters
SEQ_LENGTH = 60  # 1 hour of data (1-minute intervals)
TEST_SIZE = 0.3  # Changed to 30% for better balance

# Ensure we have enough data
min_required = SEQ_LENGTH * 2  # Minimum data needed
if len(scaled_data) < min_required:
    print(f"Error: Need at least {min_required} data points, but only have {len(scaled_data)}")
    exit()

# Split data sequentially (last 30% for testing)
split_idx = int(len(scaled_data) * (1 - TEST_SIZE))
train_data = scaled_data[:split_idx]
test_data = scaled_data[split_idx:]

print(f"Training data points: {len(train_data)}")
print(f"Test data points: {len(test_data)}")

# Create sequences
X_train = create_sequences(train_data, SEQ_LENGTH)
X_test = create_sequences(test_data, SEQ_LENGTH)

print(f"Training sequences: {X_train.shape}")
print(f"Test sequences: {X_test.shape}")

# Ensure we have enough sequences
if len(X_train) < 10 or len(X_test) < 5:
    print("Error: Not enough sequences for training/testing")
    print("Try reducing SEQ_LENGTH or using a larger dataset")
    exit()

# Build improved CNN-LSTM Autoencoder model
model = Sequential([
    # Encoder
    Conv1D(filters=64, kernel_size=3, activation='relu',
           input_shape=(SEQ_LENGTH, 1), padding='same'),
    MaxPooling1D(pool_size=2, padding='same'),
    Dropout(0.2),
    Conv1D(filters=32, kernel_size=3, activation='relu', padding='same'),
    MaxPooling1D(pool_size=2, padding='same'),
    Dropout(0.2),

    # LSTM layers for sequence learning
    LSTM(32, return_sequences=False),

    # Bottleneck
    Dense(16, activation='relu'),
    Dropout(0.2),

    # Expand back to sequence length
    RepeatVector(SEQ_LENGTH),

    # Decoder LSTM
    LSTM(32, return_sequences=True),
    Dropout(0.2),

    # TimeDistributed layer to get back to original shape
    TimeDistributed(Dense(32, activation='relu')),
    Dropout(0.2),

    # Final output layer
    TimeDistributed(Dense(1))
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
print("Model compiled successfully")
model.summary()

# Train the model
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("Starting model training...")
history = model.fit(
    X_train, X_train,  # Autoencoder: input = target
    epochs=100,
    batch_size=min(32, len(X_train)//4),  # Adjust batch size based on data size
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1,
    shuffle=True
)

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Training History - Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Model Training History - MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
plt.show()

# Make predictions and calculate reconstruction error
print("Making predictions...")
train_pred = model.predict(X_train, verbose=0)
train_mae = np.mean(np.abs(train_pred - X_train), axis=(1, 2))

test_pred = model.predict(X_test, verbose=0)
test_mae = np.mean(np.abs(test_pred - X_test), axis=(1, 2))

# Improved threshold calculation using statistical methods
# Method 1: MAD-based threshold
mad = np.median(np.abs(train_mae - np.median(train_mae)))
threshold_mad = np.median(train_mae) + 3 * mad

# Method 2: Percentile-based threshold
threshold_percentile = np.percentile(train_mae, 95)

# Method 3: Standard deviation-based threshold
threshold_std = np.mean(train_mae) + 2 * np.std(train_mae)

# Use the most conservative threshold
threshold = max(threshold_mad, threshold_percentile, threshold_std)

print(f"\nThreshold calculations:")
print(f"MAD-based threshold: {threshold_mad:.4f}")
print(f"95th percentile threshold: {threshold_percentile:.4f}")
print(f"Mean + 2*std threshold: {threshold_std:.4f}")
print(f"Final threshold (most conservative): {threshold:.4f}")

# Detect anomalies
anomalies = test_mae > threshold
anomaly_indices = np.where(anomalies)[0]

print(f"\nDetected {len(anomaly_indices)} anomalies in test data")
print(f"Anomaly percentage: {len(anomaly_indices)/len(test_mae)*100:.2f}%")

# Create ground truth using statistical methods
# Calculate ground truth based on the actual power values at the end of each sequence
power_values_test = scaler.inverse_transform(X_test[:, -1, :])  # Get last value of each sequence
power_values_test = power_values_test.flatten()

# Statistical thresholds for ground truth
z_scores = np.abs(stats.zscore(power_values_test))
q1 = np.percentile(power_values_test, 25)
q3 = np.percentile(power_values_test, 75)
iqr = q3 - q1
median_power = np.median(power_values_test)

# Create ground truth using multiple criteria
ground_truth = (
    (power_values_test < (q1 - 1.5 * iqr)) |
    (power_values_test > (q3 + 1.5 * iqr)) |
    (z_scores > 2.5)  # Z-score > 2.5 (less strict than 3)
)

print(f"Ground truth anomalies: {np.sum(ground_truth)}")
print(f"Ground truth anomaly percentage: {np.sum(ground_truth)/len(ground_truth)*100:.2f}%")

# Calculate performance metrics only if we have both normal and anomaly cases
if np.sum(ground_truth) > 0 and np.sum(~ground_truth) > 0:
    accuracy = accuracy_score(ground_truth, anomalies)
    precision = precision_score(ground_truth, anomalies, zero_division=0)
    recall = recall_score(ground_truth, anomalies, zero_division=0)
    f1 = f1_score(ground_truth, anomalies, zero_division=0)

    print(f"\n=== Performance Metrics ===")
    print(f"Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
    print(f"Precision: {precision:.4f} ({precision*100:.2f}%)")
    print(f"Recall: {recall:.4f} ({recall*100:.2f}%)")
    print(f"F1-Score: {f1:.4f} ({f1*100:.2f}%)")

    # Confusion Matrix
    cm = confusion_matrix(ground_truth, anomalies)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Anomaly'],
                yticklabels=['Normal', 'Anomaly'])
    plt.title('Confusion Matrix')
    plt.ylabel('Actual (Statistical Ground Truth)')
    plt.xlabel('Predicted')
    plt.show()

    # Print classification report
    print("\nClassification Report:")
    print(classification_report(ground_truth, anomalies, target_names=['Normal', 'Anomaly']))
else:
    print("Warning: Cannot calculate performance metrics - ground truth has only one class")
    accuracy = precision = recall = f1 = 0

# Plot results
plt.figure(figsize=(15, 10))

# Calculate time indices for plotting - Fixed indexing
start_test_idx = split_idx + SEQ_LENGTH - 1  # Adjusted for sequence alignment
test_times = df.index[start_test_idx:start_test_idx + len(power_values_test)]

# Ensure arrays have same length
min_length = min(len(test_times), len(power_values_test))
test_times = test_times[:min_length]
power_values_test = power_values_test[:min_length]
test_mae = test_mae[:min_length]

# Update anomaly indices to match the trimmed arrays
anomaly_indices = anomaly_indices[anomaly_indices < min_length]
ground_truth = ground_truth[:min_length]

print(f"Plotting arrays - Times: {len(test_times)}, Power: {len(power_values_test)}, MAE: {len(test_mae)}")

# Plot 1: Original test data with anomalies
plt.subplot(2, 2, 1)
plt.plot(test_times, power_values_test, label='Actual Power', alpha=0.7, linewidth=1)
if len(anomaly_indices) > 0:
    plt.scatter(test_times[anomaly_indices], power_values_test[anomaly_indices],
               color='red', s=30, label='Predicted Anomalies', zorder=5)
if np.sum(ground_truth) > 0:
    plt.scatter(test_times[ground_truth], power_values_test[ground_truth],
               color='green', s=50, marker='x', label='Statistical Anomalies', zorder=6)
plt.title('Power Consumption with Anomaly Detection\n(Test Data)')
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.legend()
plt.xticks(rotation=45)

# Plot 2: Reconstruction errors
plt.subplot(2, 2, 2)
plt.plot(test_times, test_mae, label='Reconstruction Error', linewidth=1)
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold ({threshold:.4f})')
if len(anomaly_indices) > 0:
    plt.scatter(test_times[anomaly_indices], test_mae[anomaly_indices],
               color='red', s=30, label='Predicted Anomalies')
plt.title('Reconstruction Error and Anomaly Detection')
plt.xlabel('Time')
plt.ylabel('Mean Absolute Error')
plt.legend()
plt.xticks(rotation=45)

# Plot 3: Distribution of errors
plt.subplot(2, 2, 3)
plt.hist(train_mae, bins=50, alpha=0.7, label='Training Errors', density=True)
plt.hist(test_mae, bins=50, alpha=0.7, label='Test Errors', density=True)
plt.axvline(x=threshold, color='r', linestyle='--', label=f'Threshold ({threshold:.4f})')
plt.title('Error Distribution')
plt.xlabel('Reconstruction Error')
plt.ylabel('Density')
plt.legend()

# Plot 4: Training vs test data split
plt.subplot(2, 2, 4)
plt.plot(df.index[:split_idx], scaler.inverse_transform(train_data),
         label='Training Data', alpha=0.7, linewidth=1)
plt.plot(df.index[split_idx:split_idx+len(test_data)], scaler.inverse_transform(test_data),
         label='Test Data', alpha=0.7, linewidth=1, color='orange')
plt.axvline(x=df.index[split_idx], color='red', linestyle='--', label='Train/Test Split')
plt.title('Train-Test Data Split')
plt.xlabel('Time')
plt.ylabel('Power (W)')
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Detailed analysis
print("\n=== Detailed Analysis ===")
print(f"Median power: {median_power:.2f}W")
print(f"IQR: {iqr:.2f}W")
print(f"Q1: {q1:.2f}W, Q3: {q3:.2f}W")
print(f"Normal range (1.5*IQR method): [{q1 - 1.5*iqr:.2f}W, {q3 + 1.5*iqr:.2f}W]")

# Analyze detection performance by power level
if len(power_values_test) > 0:
    power_min, power_max = np.min(power_values_test), np.max(power_values_test)
    power_bins = np.linspace(power_min, power_max, 6)  # Reduced bins for better analysis
    detection_stats = []

    for i in range(len(power_bins) - 1):
        low, high = power_bins[i], power_bins[i+1]
        mask = (power_values_test >= low) & (power_values_test < high)

        if np.any(mask):
            detected = np.sum(anomalies[mask])
            actual = np.sum(ground_truth[mask])
            total = np.sum(mask)
            detection_rate = detected/actual if actual > 0 else 0
            false_positive_rate = max(0, detected - actual)/max(1, total - actual) if total > actual else 0

            detection_stats.append({
                'power_range': f"{low:.0f}-{high:.0f}W",
                'detected': detected,
                'actual': actual,
                'total': total,
                'detection_rate': detection_rate,
                'false_positive_rate': false_positive_rate
            })

    print("\nDetection Performance by Power Level:")
    for stat in detection_stats:
        if stat['total'] > 0:
            print(f"{stat['power_range']}: {stat['detected']}/{stat['actual']} detected "
                  f"(Detection: {stat['detection_rate']*100:.1f}%, "
                  f"FP Rate: {stat['false_positive_rate']*100:.1f}%)")

# Improved function to detect anomalies in new data
def detect_fridge_anomalies(new_data, model, scaler, seq_length=SEQ_LENGTH, threshold=threshold):
    """
    Detect anomalies in new fridge power data

    Args:
        new_data: DataFrame with 'Power' column
        model: Trained autoencoder model
        scaler: Fitted StandardScaler
        seq_length: Length of sequences for the model
        threshold: Anomaly threshold

    Returns:
        anomalies: Boolean array indicating anomalies
        point_errors: Reconstruction errors for each point
        anomaly_count: Total number of anomalies
    """
    if len(new_data) < seq_length:
        raise ValueError(f"Need at least {seq_length} data points, got {len(new_data)}")

    # Ensure 'Power' column exists
    if 'Power' not in new_data.columns:
        raise ValueError("'Power' column not found in new_data")

    scaled_new = scaler.transform(new_data[['Power']])
    X_new = create_sequences(scaled_new, seq_length)

    if len(X_new) == 0:
        raise ValueError("No sequences could be created from the data")

    predictions = model.predict(X_new, verbose=0)
    errors = np.mean(np.abs(predictions - X_new), axis=(1, 2))

    anomalies = errors > threshold
    anomaly_count = np.sum(anomalies)

    # Align errors with original data points
    point_errors = np.full(len(new_data), np.nan)
    point_errors[seq_length-1:seq_length-1+len(errors)] = errors

    return anomalies, point_errors, anomaly_count

# Test the function with proper error handling
try:
    # Create test subset that aligns with our sequences
    test_start_idx = split_idx
    test_end_idx = split_idx + len(test_data)
    test_df_subset = df.iloc[test_start_idx:test_end_idx].copy()

    if len(test_df_subset) >= SEQ_LENGTH:
        anomalies_result, errors_result, count = detect_fridge_anomalies(test_df_subset, model, scaler)

    print(f"\n=== Function Test Results ===")
    print(f"Total test sequences: {len(X_test)}")
    print(f"Detected anomalies: {count}")
    print(f"Anomaly rate: {count/len(X_test)*100:.2f}%")
except Exception as e:
    print(f"Error testing detection function: {e}")

# Save the model and scaler for future use
try:
    model.save('fridge_anomaly_detector.h5')

    # Save scaler parameters
    import joblib
    joblib.dump(scaler, 'fridge_scaler.pkl')

    # Save threshold and parameters
    np.savez('model_parameters.npz',
             threshold=threshold,
             seq_length=SEQ_LENGTH,
             median_power=median_power,
             iqr=iqr)

    print("Model and parameters saved successfully!")
    print("Files created:")
    print("- fridge_anomaly_detector.h5 (model)")
    print("- fridge_scaler.pkl (scaler)")
    print("- model_parameters.npz (threshold and parameters)")
except Exception as e:
    print(f"Error saving model: {e}")

# Additional pattern analysis
print("\n=== Pattern Analysis ===")
if len(X_train) > 0:
    normal_cycles = np.mean(X_train, axis=0).flatten()
    print("Average normal cycle pattern (first 10 minutes):")
    for i in range(min(10, len(normal_cycles))):
        power_val = scaler.inverse_transform([[normal_cycles[i]]])[0][0]
        print(f"Minute {i+1}: {power_val:.2f}W")

if len(anomaly_indices) > 0:
    anomaly_power_levels = []
    for idx in anomaly_indices:
        if idx < len(X_test):
            avg_power = np.mean(scaler.inverse_transform(X_test[idx]))
            anomaly_power_levels.append(avg_power)

    if anomaly_power_levels:
        print(f"\nAverage power during anomalies: {np.mean(anomaly_power_levels):.2f}W")
        print(f"Range of anomaly power: {np.min(anomaly_power_levels):.2f}W to {np.max(anomaly_power_levels):.2f}W")

# Summary statistics
print(f"\n=== Summary Statistics ===")
print(f"Dataset size: {len(df)} data points")
print(f"Training sequences: {len(X_train)}")
print(f"Test sequences: {len(X_test)}")
print(f"Sequence length: {SEQ_LENGTH} minutes")
print(f"Anomaly threshold: {threshold:.4f}")
if accuracy > 0:
    print(f"Detection accuracy: {accuracy*100:.2f}%")
    print(f"Precision: {precision*100:.2f}%")
    print(f"Recall: {recall*100:.2f}%")
print(f"Model training epochs: {len(history.history['loss'])}")
print(f"Final training loss: {history.history['loss'][-1]:.4f}")
print(f"Final validation loss: {history.history['val_loss'][-1]:.4f}")

print("\n=== Usage Instructions ===")
print("To use the saved model for new data:")
print("1. Load the model: model = tf.keras.models.load_model('fridge_anomaly_detector.h5')")
print("2. Load the scaler: scaler = joblib.load('fridge_scaler.pkl')")
print("3. Load parameters: params = np.load('model_parameters.npz')")
print("4. Use detect_fridge_anomalies(new_data, model, scaler) function")
```
